3) Next we import the csv reader alongside the Beautifulsoup library for the purpose of data scraping from the wikipedia page listed in the assignment.

```python
from bs4 import BeautifulSoup
import requests
from pandas.io.html import read_html
import csv
import pandas as pd
```

```python
source=requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source,'lxml')
#print(soup.prettify())
```

4) To be able to scrap the data in the needed table we will generate a for loop and save it in a csv file which we will transform into a DataFrame to be able to wrangle and work with the data within.

```python
csv_file=open('cms_scrape.csv','w')
csv_writer=csv.writer(csv_file)
csv_writer.writerow(['Postal Code','Borough','Neighborhood'])
for td in soup.find_all('td'):
    #print(td)
    print()
    pc=td.b.text
    print(pc)
    try:
        b=td.p.a.text
    except Exception as e:
        b="Not assigned"
    print(b)
    try:
        n=td.p.text
    except Exception as e:
        pass
    print(n)
    
csv_writer.writerow([pc,b,n])
csv_file.close()
```

5) Reading our csv file we get:

```python
data = pd.read_csv("cms_scrape.csv", delimiter=",")
data.head()
```

6) Below we find the first 5 rows of our data frame.
